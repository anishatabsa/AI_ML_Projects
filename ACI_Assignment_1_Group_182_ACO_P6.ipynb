```python
import numpy as np
import random
import matplotlib.pyplot as plt
from copy import deepcopy
```

```python
weights = []
values = []
capacity = 0
n_items = 0
n_ants = 0
alpha = 0
beta = 0
evaporation_rate = 0
max_iterations = 100
best_solution = None
best_value = 0
convergence_history = []
pheromones = []
heuristic = 0.0
```

```python
#Code Block: Set Initial State (Must handle dynamic inputs)
def set_initial_state():
    global weights, values, capacity, n_items, n_ants, alpha, beta, evaporation_rate, max_iterations, best_solution, best_value, convergence_history, pheromones, heuristic
    input_weights = input("Enter 10 weights (comma separated) : ").split(',')
    input_values = input("Enter 10 values (comma separated) : ").split(',')
    capacity = int(input("Enter capacity value: "))
    weights = np.array(input_weights)
    values = np.array(input_values)
    capacity = capacity
    n_items = len(weights)
    n_ants = n_ants
    alpha = alpha  # Pheromone importance
    beta = beta    # Heuristic importance (distance/desirability)
    evaporation_rate = evaporation_rate
    max_iterations = max_iterations
```

```python
#Code Block: Set the matrix for transition & cost (as relevant for the given problem)
# Initialize pheromone matrix (items x position in selection order)
def initialize_pheromones():
    global pheromones, heuristic, n_items, values, weights
    pheromones = np.ones((n_items, n_items)) * 0.1

    # Calculate heuristic information (value/weight ratio)
    heuristic = values / weights

```

```python
#Code Block : Write function to design the Transition Model/Successor function. Ideally this would be called while search algorithms are implemented
def construct_solution(self, ant_id):
    """Construct a solution for one ant"""
    solution = np.zeros(self.n_items, dtype=int)
    remaining_capacity = self.capacity
    available_items = list(range(self.n_items))
    selection_order = []

    while available_items and remaining_capacity > 0:
        # Calculate probabilities for available items
        probabilities = self.calculate_probabilities(available_items, remaining_capacity)

        if not probabilities:
            break

        # Select item based on probabilities
        selected_idx = np.random.choice(len(available_items), p=probabilities)
        selected_item = available_items[selected_idx]

        # Add item if it fits
        if self.weights[selected_item] <= remaining_capacity:
            solution[selected_item] = 1
            remaining_capacity -= self.weights[selected_item]
            selection_order.append(selected_item)

        # Remove selected item from available items
        available_items.remove(selected_item)

    return solution, selection_order


def calculate_probabilities(available_items, remaining_capacity):
    """Calculate selection probabilities for available items"""
    global pheromones, weights, values, n_ants, max_iterations, alpha, beta, evaporation_rate

    if not available_items:
        return []

    # Filter items that can fit
    feasible_items = [item for item in available_items if weights[item] <= remaining_capacity]

    if not feasible_items:
        return []

    # Calculate probability components
    probabilities = []
    for item in feasible_items:
        # Pheromone component (average pheromone for this item)
        pheromone_value = np.mean(pheromones[item, :])

        # Heuristic component (value/weight ratio)
        heuristic_value = heuristic[item]

        # Combined probability
        prob = (pheromone_value ** alpha) * (heuristic_value ** beta)
        probabilities.append(prob)

    # Normalize probabilities
    total_prob = sum(probabilities)
    if total_prob > 0:
        probabilities = [p / total_prob for p in probabilities]
    else:
        probabilities = [1.0 / len(feasible_items)] * len(feasible_items)

    # Map back to available_items indices
    full_probabilities = []
    feasible_idx = 0
    for item in available_items:
        if item in feasible_items:
            full_probabilities.append(probabilities[feasible_idx])
            feasible_idx += 1
        else:
            full_probabilities.append(0.0)

    # Normalize again
    total_prob = sum(full_probabilities)
    if total_prob > 0:
        full_probabilities = [p / total_prob for p in full_probabilities]

    return full_probabilities
```

```python
def update_pheromones(solutions, values):
    """Update pheromone trails based on ant solutions"""
    # Evaporation
    global pheromones
    pheromones *= (1 - evaporation_rate)

    # Add pheromones based on solution quality
    for i, (solution, selection_order) in enumerate(solutions):
        if values[i] > 0:  # Only update for feasible solutions
            # Pheromone deposit proportional to solution quality
            deposit = values[i] / best_value if best_value > 0 else 1.0

            # Update pheromones for selected items
            for pos, item in enumerate(selection_order):
                pheromones[item, pos] += deposit
```

```python
def optimize():
    """Main ACO optimization loop"""
    global best_solution, best_value, convergence_history
    print(f"Starting ACO with {n_ants} ants, alpha={alpha}, beta={beta}")
    print("=" * 60)

    for iteration in range(max_iterations):
        # Generate solutions for all ants
        solutions = []
        values = []

        for ant in range(n_ants):
            solution, selection_order = construct_solution(ant)
            value = evaluate_solution(solution)
            solutions.append((solution, selection_order))
            values.append(value)

            # Update best solution
            if value > best_value:
                best_value = value
                best_solution = solution.copy()

        # Update pheromones
        update_pheromones(solutions, values)

        # Track convergence
        convergence_history.append(best_value)

        # Print progress
        if iteration % 10 == 0 or iteration == max_iterations - 1:
            avg_value = np.mean(values)
            print(f"Iteration {iteration:3d}: Best = {best_value:2d}, Average = {avg_value:5.2f}")

    return best_solution, best_value
```

```python
#Code block: Write fucntion to handle goal test (Must handle dynamic inputs). Ideally this would be called while search algorithms are implemented
def evaluate_solution(self, solution):
    """Evaluate the quality of a solution"""
    total_weight = np.sum(solution * self.weights)
    total_value = np.sum(solution * self.values)

    if total_weight <= self.capacity:
        return total_value
    else:
        return 0  # Infeasible solution
```

```python
def print_results(scenario_name):
    """Print detailed results"""
    global best_solution, best_value, convergence_history
    print(f"\n=== {scenario_name} RESULTS ===")
    if best_solution is not None:
        selected_items = [i+1 for i, selected in enumerate(best_solution) if selected]
        total_weight = np.sum(best_solution * weights)

        print(f"Best Value: {best_value}")
        print(f"Total Weight: {total_weight}/{capacity} kg")
        print(f"Selected Items: {selected_items}")
        print(f"Item Details:")

        for i, selected in enumerate(best_solution):
            if selected:
                print(f"  Item {i+1}: Weight={weights[i]}, Value={values[i]}, "
                      f"Efficiency={values[i]/weights[i]:.2f}")

    print(f"Convergence: Reached best value at iteration {len(convergence_history) - 1 - convergence_history[::-1].index(best_value)}")
```
